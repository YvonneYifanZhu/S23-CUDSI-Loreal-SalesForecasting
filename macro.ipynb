```python
import pandas as pd
import numpy as np
```

```python
macro = pd.read_excel("macro_indicators.xlsx")
traffic = pd.read_csv("traffic_cleaned_forPython.csv")
macro = macro.rename(columns={'Date': 'date'})
traffic = traffic.rename(columns={'day': 'date'})
```

```python
traffic["date"] = pd.to_datetime(traffic["date"])
macro["date"] = pd.to_datetime(macro["date"])
macro = macro[(macro["date"]>=pd.to_datetime('2015-04')) & (macro["date"]<=pd.to_datetime('2019-09'))]
traffic['year_month'] = traffic['date'].dt.strftime('%Y-%m')
macro['year_month'] = macro['date'].dt.strftime('%Y-%m')
```

```python
joined_df = pd.merge(traffic, macro, on='year_month')
joined_df
```

```python
joined_df.corr()
```

```python
month_traffic = traffic.groupby(pd.Grouper(key='date', freq='M')).mean()
month_macro = macro.groupby(pd.Grouper(key='date', freq='M')).mean()
concatenated = pd.concat([month_traffic, month_macro], axis=1)
cov = concatenated.corr()
cov
```

```python
## A noisy sine wave as query
idx = np.linspace(0,6.28,num=100)
query = np.sin(idx) + np.random.uniform(size=100)/10.0

## A cosine is for template; sin and cos are offset by 25 samples
template = np.cos(idx)

## Find the best match with the canonical recursion formula
from dtw import *
alignment = dtw(joined_df["CORE_VL_NbEntry"], joined_df["US ISM Manufacturing New Orders Index"], keep_internals=True)

## Align and plot with the Rabiner-Juang type VI-c unsmoothed recursion
dtw(query, template, keep_internals=True, 
    step_pattern=rabinerJuangStepPattern(6, "c"))\
    .plot(type="twoway",offset=-2)
```
